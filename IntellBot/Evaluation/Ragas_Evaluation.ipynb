```python
import textwrap
import re  # Added import for regular expressions

def wraptextnewlines(text, width=110):
    lines = text.split('\n')
    wrapped = [textwrap.fill(line, width=width) for line in lines]
    wrappedtext = '\n'.join(wrapped)
    return wrappedtext

def response(llm_response):
    result = wraptextnewlines(llm_response['result'])
    print(result)

    # Use regular expression to match similar "I don't know" variants
    if not any(re.search(phrase, result, re.IGNORECASE) for phrase in ["I don't know", "not mentioned", "no information"]):
        print("\n\nSources:")
        seen_sources = set()  # Set to store unique sources
        for s in llm_response["source_documents"]:
            source_name = s.metadata['source']
            if source_name not in seen_sources:
                print(source_name)
                seen_sources.add(source_name)  # Add the source to the set of seen sources
```
