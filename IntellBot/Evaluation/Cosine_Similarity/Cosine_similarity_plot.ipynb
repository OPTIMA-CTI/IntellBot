```python
import matplotlib.pyplot as plt

# Sample data for three sets of values
categories = ['Vulnerability', 'APT', 'Security Blog', 'VirusTotal Report']
values_word2vec = [0.847, 0.836, 0.806, 0.987]
values_glove = [0.885, 0.870, 0.828, 0.891]
values_bert = [0.916, 0.9, 0.918, 0.872]

# Increasing figure size
plt.figure(figsize=(16, 14))

# Plotting the data
bar_width = 0.25
index = range(len(categories))

# Creating bars for each set of values
plt.bar([i for i in index], values_word2vec, bar_width, label='Word2Vec', color='black')
plt.bar([i + bar_width for i in index], values_glove, bar_width, label='GloVe', color='white', edgecolor='black', hatch='//')
plt.bar([i + 2 * bar_width for i in index], values_bert, bar_width, label='BERT', color='grey',edgecolor='black')

# Adding data labels
for i, v in enumerate(values_word2vec):
    plt.text(i, v + 0.005, str(v), ha='center', va='bottom', color='black', fontsize=20, fontweight='bold')
for i, v in enumerate(values_glove):
    plt.text(i + bar_width, v + 0.005, str(v), ha='center', va='bottom', color='black', fontsize=20, fontweight='bold')
for i, v in enumerate(values_bert):
    plt.text(i + 2 * bar_width, v + 0.005, str(v), ha='center', va='bottom', color='black', fontsize=20, fontweight='bold')

# Adding legend
plt.legend(fontsize=18)

# Adding labels and title with increased size and bold text
plt.xlabel('Type of Document', fontsize=26, fontweight='bold')
plt.ylabel('Cosine Similarity', fontsize=26, fontweight='bold')

# Adjusting x-axis ticks and labels with increased size and bold text
plt.xticks([i + bar_width for i in index], categories, fontsize=20, fontweight='bold')

# Adjusting y-axis ticks and labels with increased size and bold text
plt.yticks(fontsize=20, fontweight='bold')

# Adjusting y-axis range
plt.ylim(0.75, 1.00)

# Displaying the plot
plt.tight_layout()
plt.show()
```
